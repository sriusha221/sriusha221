```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from datetime import datetime, timedelta

# Sample data of subjects and recommended study hours
subjects_data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Geography'],
    'Importance': [5, 4, 3, 2, 1],  # Importance of subject (1 - lowest, 5 - highest)
    'Duration': [2, 1.5, 1, 2, 1],  # Recommended hours for each subject
}

# Convert data into a DataFrame
subjects_df = pd.DataFrame(subjects_data)

# Function to suggest a study plan based on available time and subject preferences
def create_study_plan(total_study_time, preferences):
    # Encode the preferences into numerical values
    encoder = LabelEncoder()
    preferences_encoded = encoder.fit_transform(preferences)

    # Add preferences to the subjects DataFrame
    subjects_df['Preference'] = preferences_encoded

    # Sort subjects by importance and preference
    subjects_df['WeightedImportance'] = subjects_df['Importance'] * subjects_df['Preference']
    subjects_df_sorted = subjects_df.sort_values(by='WeightedImportance', ascending=False)

    # Calculate study plan based on available time
    study_plan = []
    remaining_time = total_study_time

    for index, row in subjects_df_sorted.iterrows():
        if remaining_time <= 0:
            break

        subject_time = min(row['Duration'], remaining_time)
        study_plan.append({
            'Subject': row['Subject'],
            'Time Allocated (hrs)': subject_time
        })

        remaining_time -= subject_time

    return pd.DataFrame(study_plan)

# Sample Input: Available study time and preferences
available_time = float(input("Enter available study time in hours: "))
preferences = input("Enter your subject preferences (comma-separated, e.g., Math, Science, English): ").split(',')

# Clean up the preferences input
preferences = [pref.strip() for pref in preferences]

# Create the study plan
study_plan_df = create_study_plan(available_time, preferences)

# Display the suggested study plan
print("\nSuggested Study Plan:")
print(study_plan_df)
```

```python
import datetime
import random
import json

# Function to get user input
def get_user_input():
    subjects = input("Enter your subjects/topics separated by commas: ").split(',')
    study_hours = int(input("How many hours per day can you study? "))
    days_available = int(input("How many days do you have before the exam? "))
    priority_subject = input("Which subject is the most important for you? ")

    return {
        "subjects": [subject.strip() for subject in subjects],
        "study_hours": study_hours,
        "days_available": days_available,
        "priority_subject": priority_subject.strip()
    }

# Function to generate study plan
def generate_study_plan(user_data):
    subjects = user_data["subjects"]
    total_days = user_data["days_available"]
    daily_hours = user_data["study_hours"]
    priority_subject = user_data["priority_subject"]

    study_plan = {}

    for day in range(1, total_days + 1):
        date = datetime.date.today() + datetime.timedelta(days=day)
        study_plan[str(date)] = {}

        # Allocate more time to priority subject
        priority_time = round(daily_hours * 0.4, 2)  # 40% for priority subject
        remaining_time = daily_hours - priority_time

        study_plan[str(date)][priority_subject] = f"{priority_time} hours"

        # Allocate remaining time to other subjects
        other_subjects = [sub for sub in subjects if sub != priority_subject]
        random.shuffle(other_subjects)

        if other_subjects:  # Ensure there's at least one other subject
            for subject in other_subjects:
                allocated_time = round(remaining_time / len(other_subjects), 2)
                study_plan[str(date)][subject] = f"{allocated_time} hours"
        else:
            study_plan[str(date)][priority_subject] = f"{daily_hours} hours"  # Assign all time if no other subjects

    return study_plan

# Function to display study plan
def display_study_plan(study_plan):
    print("\nGenerated Study Plan:")
    for date, schedule in study_plan.items():
        print(f"\nðŸ“… Date: {date}")
        for subject, hours in schedule.items():
            print(f"  ðŸ“– {subject}: {hours}")

# Main function
def main():
    print("ðŸ“š Welcome to the AI Study Planner ðŸ“š")
    user_data = get_user_input()
    study_plan = generate_study_plan(user_data)

    # Save study plan as JSON
    with open("study_plan.json", "w") as file:
        json.dump(study_plan, file, indent=4)

    display_study_plan(study_plan)
    print("\nâœ… Your study plan has been saved as 'study_plan.json'.")

# Corrected main execution check
if __name__ == "__main__":
    main()
```

```python
import random
import datetime

class StudyPlanner:
    def __init__(self, user_name, subjects, learning_style):
        self.user_name = user_name
        self.subjects = subjects
        self.learning_style = learning_style
        self.schedule = {}

    def generate_schedule(self):
        # Randomly assign subjects to time slots
        current_time = datetime.datetime.now()
        for subject in self.subjects:
            study_time = random.randint(1,2)  # Study duration 1-2 hours
            study_time_slot = current_time + datetime.timedelta(hours=study_time)
            self.schedule[subject] = study_time_slot
            current_time = study_time_slot

    def show_schedule(self):
        print(f"Study Schedule for {self.user_name}:")
        for subject, time in self.schedule.items():
            print(f"{subject}: {time.strftime('%Y-%m-%d %H:%M:%S')}")

    def get_personalized_recommendations(self):
        recommendations = []
        if self.learning_style == "visual":
            recommendations.append("Watch video tutorials")
        elif self.learning_style == "auditory":
            recommendations.append("Listen to podcasts")
        else:
            recommendations.append("Use textbooks and written notes")
        return recommendations

subjects = ['Math', 'Science', 'History']
planner = StudyPlanner('John', subjects, 'visual')
planner.generate_schedule()
planner.show_schedule()
print(planner.get_personalized_recommendations())
```

```python
class AIStudyPlanner:
    def __init__(self, subjects, hours_per_day=4): # Changed _init_ to __init__
        if not isinstance(subjects, list) or len(subjects) == 0:
            raise ValueError("Subjects should be a non-empty list.")
        if not isinstance(hours_per_day, (int, float)) or hours_per_day <= 0:
            raise ValueError("Hours per day should be a positive number.")

        self.subjects = subjects
        self.hours_per_day = hours_per_day
        self.schedule = []

    def generate_schedule(self, days=7):
        """Generate a study schedule with improved error handling."""
        if days <= 0:
            raise ValueError("Days should be a positive number.")

        for day in range(days):
            daily_plan = []
            random.shuffle(self.subjects)
            for subject in self.subjects:
                hours = round(self.hours_per_day / len(self.subjects), 2)
                daily_plan.append({"Day": day + 1, "Subject": subject, "Hours": hours})
            self.schedule.append(daily_plan)
        return self.schedule

    def display_schedule(self):
        """Display the study schedule in a structured way."""
        for day_plan in self.schedule:
            print(f"\nDay {day_plan[0]['Day']} Study Plan:")
            for task in day_plan:
                print(f"  - {task['Subject']}: {task['Hours']} hours")
```

```python
# Install necessary libraries (Run this in Google Colab)
!pip install openai pandas matplotlib tensorflow

import os
import openai
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np

# Set OpenAI API Key securely
openai.api_key = os.getenv("OPENAI_API_KEY")  # Ensure you set this in your environment

# Function to generate study plan using OpenAI GPT
def generate_study_plan(subjects, hours_per_day):
    subject_list = ", ".join(subjects)  # Convert list to a string
    prompt = f"Create a study schedule for the following subjects: {subject_list}. Allocate {hours_per_day} hours per day efficiently."

    try:
        response = openai.ChatCompletion.create(
            model="gpt-4",
            messages=[{"role": "system", "content": "You are an AI study planner."},
                      {"role": "user", "content": prompt}]
        )
        return response['choices'][0]['message']['content']
    except Exception as e:
        return f"Error fetching study plan: {e}"

# Example Subjects
subjects = ["Math", "Physics", "Computer Science", "AI"]
hours_per_day = 6

# Generate Study Plan
study_plan = generate_study_plan(subjects, hours_per_day)
print("Generated Study Plan:\n", study_plan)

# Data Preparation for Optimization
study_data = pd.DataFrame({
    'Subject': subjects,
    'Priority': np.random.randint(1, 10, size=len(subjects)),  # Random priority scores
    'Hours Allocated': [hours_per_day / len(subjects)] * len(subjects)  # Equal distribution
})

# Normalize Priority Scores
study_data['Priority'] = study_data['Priority'] / study_data['Priority'].sum()

# Optimize Time Allocation Using TensorFlow (Basic Example)
def optimize_study_hours(data):
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(1, input_shape=[1], activation='relu')
    ])

    model.compile(optimizer='adam', loss='mse')

    # Ensure input is reshaped correctly
    X = data['Priority'].values.reshape(-1, 1)  # Convert (n,) to (n,1)
    Y = data['Hours Allocated'].values.reshape(-1, 1)  # Convert (n,) to (n,1)

    model.fit(X, Y, epochs=100, verbose=1)  # Keep verbose=1 for minimal output

    optimized_hours = model.predict(X).flatten()  # Flatten from (n,1) to (n,)
    data['Optimized Hours'] = optimized_hours
    return data

# Apply AI Optimization
optimized_study_plan = optimize_study_hours(study_data)
print("\nOptimized Study Plan:\n", optimized_study_plan)

# Visualization
plt.figure(figsize=(10, 5))
plt.bar(optimized_study_plan['Subject'], optimized_study_plan['Optimized Hours'], color='blue')
plt.xlabel('Subjects')
plt.ylabel('Optimized Study Hours')
plt.title('AI-Optimized Study Plan')
plt.show()
```

```python
!pip install pandas numpy openai ortools

import pandas as pd
import numpy as np
import os
import openai
from ortools.sat.python import cp_model
from ortools.sat.python.cp_model import LinearExpr

# Set OpenAI API Key securely
openai.api_key = os.getenv("OPENAI_API_KEY")

# Define subjects and difficulty levels
study_data = {
    "Subject": ["Math", "Science", "History", "Programming", "Physics"],
    "Priority": [5, 4, 3, 5, 4],  # Scale from 1 to 5
    "Difficulty": [4, 3, 2, 5, 4],  # Scale from 1 to 5
    "Hours Required": [10, 8, 6, 12, 9]
}

df = pd.DataFrame(study_data)

# Define available study hours per day
available_hours = {
    "Monday": 3,
    "Tuesday": 2,
    "Wednesday": 4,
    "Thursday": 3,
    "Friday": 5,
    "Saturday": 6,
    "Sunday": 4
}

def get_study_recommendation(subject):
    """Generates a study plan using OpenAI."""
    prompt = f"Give me a brief study plan for {subject} covering key concepts and focus areas."
    try:
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[{"role": "user", "content": prompt}]
        )
        return response["choices"][0]["message"]["content"]
    except Exception as e:
        return f"Error getting study plan: {e}"

# Get AI-generated study recommendations
for subject in df["Subject"]:
    print(f"{subject} Study Plan: {get_study_recommendation(subject)}")

def create_study_schedule(df, available_hours):
    """Creates an optimized study schedule using OR-Tools."""
    model = cp_model.CpModel()
    days = list(available_hours.keys())
    subjects = df["Subject"].tolist()

    # Decision Variables
    study_hours = {
        (subject, day): model.NewIntVar(0, available_hours[day], f"{subject}_{day}")
        for subject in subjects for day in days
    }

    # Constraints: Assign required hours to subjects
    for subject in subjects:
        total_hours = sum(study_hours[(subject, day)] for day in days)
        model.Add(total_hours == df.loc[df["Subject"] == subject, "Hours Required"].values[0])

    # Constraints: Daily available hours
    for day in days:
        total_study = sum(study_hours[(subject, day)] for subject in subjects)
        model.Add(total_study <= available_hours[day])

    # Objective Function: Balance difficulty and study hours
    model.Minimize(LinearExpr.WeightedSum(
        [study_hours[(subject, day)] for subject in subjects for day in days],
        [df.loc[df["Subject"] == subject, "Difficulty"].values[0] for subject in subjects for day in days]
    ))

    # Solve
    solver = cp_model.CpSolver()
    status = solver.Solve(model)

    if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
        schedule = {day: {subject: solver.Value(study_hours[(subject, day)]) for subject in subjects} for day in days}
        return schedule
    else:
        return "No feasible solution found."

# Generate optimized study plan
study_schedule = create_study_schedule(df, available_hours)

# Print the study schedule
if isinstance(study_schedule, dict):
    study_schedule_df = pd.DataFrame(study_schedule).T  # Transpose for readability
    print(study_schedule_df)
else:
    print(study_schedule)
```

```python
import datetime
import random
import ipywidgets as widgets
from IPython.display import display

# Sample study topics
study_topics = [
    "Machine Learning Basics",
    "Deep Learning",
    "Data Structures",
    "Algorithms",
    "Mathematics for AI",
    "Python Programming",
    "Statistics",
    "Natural Language Processing",
]

# User inputs
name = widgets.Text(description="Name:")
available_hours = widgets.IntSlider(min=1, max=10, description="Daily Hours:")
difficulty_level = widgets.Dropdown(
    options=["Beginner", "Intermediate", "Advanced"], description="Level:"
)
start_date = widgets.DatePicker(description="Start Date:")

# Display inputs
display(name, available_hours, difficulty_level, start_date)

# Function to generate study plan
def generate_study_plan(name, hours, level, start):
    if not name or not start:
        print("Please enter your name and a valid start date.")
        return

    plan = []
    random.shuffle(study_topics)  # Randomize topics
    difficulty_weights = {"Beginner": 2, "Intermediate": 3, "Advanced": 4}
    study_days = difficulty_weights[level] * len(study_topics)

    for i in range(study_days):
        topic = study_topics[i % len(study_topics)]
        date = start + datetime.timedelta(days=i)
        plan.append((date.strftime("%Y-%m-%d"), topic, hours))

    print(f"\nStudy Plan for {name}:\n")
    for date, topic, hrs in plan:
        print(f"{date}: Study {topic} for {hrs} hours")

# Button to generate plan
button = widgets.Button(description="Generate Plan")

def on_button_click(b):
    generate_study_plan(name.value, available_hours.value, difficulty_level.value, start_date.value)

button.on_click(on_button_click)
display(button)
```

```python
import time

class AIStudyPlanner:
    def __init__(self):  # âœ… Fixed __init__ method
        self.default_plan = {
            "Math": ["Algebra", "Calculus"],
            "Science": ["Physics", "Biology"],
            "Programming": ["Python", "Machine Learning"]
        }

    def generate_plan(self, subject):
        """Generates a study plan based on the given subject."""
        if subject in self.default_plan:
            return {subject: self.default_plan[subject]}
        else:
            return {"Error": "Subject not found"}

    def add_custom_plan(self, subject, topics):
        """Allows users to add a custom study plan."""
        if isinstance(topics, list) and topics:  # âœ… Fixed type check
            self.default_plan[subject] = topics
            return {"Success": f"Custom plan for {subject} added"}
        return {"Error": "Invalid topics"}

# Test Script (Save this in `test_study_planner.py`)
import pytest

@pytest.fixture
def planner():
    return AIStudyPlanner()

def test_generate_valid_plan(planner):
    result = planner.generate_plan("Math")
    assert "Math" in result
    assert isinstance(result["Math"], list)

def test_generate_invalid_plan(planner):
    result = planner.generate_plan("History")
    assert "Error" in result
    assert result["Error"] == "Subject not found"

def test_add_custom_plan(planner):
    result = planner.add_custom_plan("History", ["World Wars", "Ancient Civilizations"])
    assert "Success" in result
    assert planner.generate_plan("History") == {"History": ["World Wars", "Ancient Civilizations"]}

def test_add_invalid_custom_plan(planner):
    result = planner.add_custom_plan("History", "Not a List")
    assert "Error" in result

if __name__ == "__main__":
    pytest.main(["-v"])  # âœ… Proper way to run pytest inside a script

# Performance Testing
import timeit

planner = AIStudyPlanner()

# Measure time for valid subject
valid_time = timeit.timeit(lambda: planner.generate_plan("Math"), number=1000)
print(f"Time taken for valid study plan generation: {valid_time:.6f} seconds")

# Measure time for invalid subject
invalid_time = timeit.timeit(lambda: planner.generate_plan("History"), number=1000)
print(f"Time taken for invalid study plan request: {invalid_time:.6f} seconds")
```
